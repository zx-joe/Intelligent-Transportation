```python
def pasteImg(target_img_i, background):
    
    random_angle = np.random.randint(-30, 30)

    mask = Image.new('L', target_img_i.size, 255)
    target_img_i = target_img.rotate(random_angle,expand=True)
    mask = mask.rotate(random_angle, expand=True)


    background.paste(target_img_i, (100,100),mask)
    return background
```
